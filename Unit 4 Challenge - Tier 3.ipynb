**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2020 to the price in 2000.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 2000 and 2020.
4. Calculate the ratio of the Average Price for 2000 divided by the Average Price for 2020.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# average prices in 2000
a = dfg['Average_Price'][dfg['Year']==2000]
a.head()
```

```python
# average prices in 2020
b = dfg['Average_Price'][dfg['Year']==2020]
b.head()
```

```python
# last 20 years are 2000 to 2020, need to find the biggest increase in price for each borough
def create_price_ratio(data):
    avePrice2000 = float(data['Average_Price'][data['Year']==2000])
    avePrice2020 = float(data['Average_Price'][data['Year']==2020])
    ratio = [avePrice2000/avePrice2020]
    return ratio
```

```python
dfg[dfg['London_Borough']=='Barking & Dagenham'].head()
```

```python
# testing the function
create_price_ratio(dfg[dfg['London_Borough']=='Barking & Dagenham'])
```

```python
# Need to find the ratios for all the London boroughs
# Creating a dictionary to save all the ratios
priceRatios = {}
```

```python
# iterate through the price ratios for all the boroughs
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg['London_Borough']==b]
    priceRatios[b] = create_price_ratio(borough)
```

```python
priceRatios
```

```python
# converting to a dataframe
df_price_ratios = pd.DataFrame(priceRatios)
df_price_ratios
```

```python
df_price_ratios_T = df_price_ratios.T
df_price_ratios_T.head(10)
```

```python
df_price_ratios_T = df_price_ratios_T.reset_index()
df_price_ratios_T.head()
```

```python
df_price_ratios = df_price_ratios_T.rename(columns={'index':'London_Borough', 0:'Price_Change_Ratio'})
df_price_ratios.head()
```

```python
# finding boroughs with the greatest increase
# top 10
top10 = df_price_ratios.sort_values(by='Price_Change_Ratio', ascending=False).head(10)
top10
```

```python
ax = top10[['London_Borough','Price_Change_Ratio']].plot(kind='bar')
ax.set_xticklabels(top10.London_Borough)
```
